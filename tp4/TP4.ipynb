Let's build a face recognition system. We are going to compare several models:

- Logistic regression
- SVM classifier

```python
from sklearn.datasets import fetch_lfw_people
```

```python
import matplotlib.pyplot as plt

faces = fetch_lfw_people(min_faces_per_person=70)

for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(faces.images[i])
    plt.xlabel(faces.target_names[faces.target[i]])
```

```python
print(faces.target[1])  # The output is a number
```

```python
from sklearn.model_selection import train_test_split

X = faces.data
y = faces.target
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```
